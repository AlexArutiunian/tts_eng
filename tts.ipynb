```python
# import all the modules that we will need to use
from TTS.utils.manage import ModelManager
from TTS.utils.synthesizer import Synthesizer

path = "/usr/local/lib/python3.10/dist-packages/TTS/.models.json"

model_manager = ModelManager(path)

# tts_models/en/male_voice_model

model_path, config_path, model_item = model_manager.download_model("tts_models/en/ljspeech/tacotron2-DDC_ph")

voc_path, voc_config_path, _ = model_manager.download_model(model_item["default_vocoder"])

syn = Synthesizer(
    tts_checkpoint=model_path,
    tts_config_path=config_path,
    vocoder_checkpoint=voc_path,
    vocoder_config=voc_config_path
)


text = """In order to make this information available to as wide a group as possible, I have avoided technical
language and will ask the indulgence of professional readers. """


def TTS(file_name):
    filename = file_name

    with open(rf"{filename}", "r", encoding="utf-8") as f:
        text = f.read()

    outputs = syn.tts(text)
    syn.save_wav(outputs, fr"{filename}.wav")



TTS(rf"text.txt")
```
